## Missing Values 

```python
## Here we will check the percentage of nan values present in each features

## 1st step make the list of all the features which has missing values
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

## 2nd step print the feature name and the percentage of missing values

for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

## Since there many missing values. So now we need to find the relationship between missing values and Sales price if any

```python
for feautre in features_with_nan:
    data=dataset.copy()
    
    ## lets make a varibale which indicates 1 if the observation is missing and 0 if there is an observation
    data[feature]=np.where(data[feature].isnull(), 1, 0)
    
    ## Now we calculate the salesPrice mean where observation is present or absent
    ## Median is calculate of the sale price with respect to feature 
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feautre)
    plt.show()
```

Form the above graph we can see that the nan values are making an impact also. So we have to replace them with the some significant values

## Numerical Variables 

```python
## list of numerical variables
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes != 'O']

print('Number of numerical features: ', len(numerical_features))
```

```python
## numerical features
dataset[numerical_features].head()
```

### Temporal Variables (DateTime Variables)

Here we have 4 temporal variables using which we can find the difference between when the house was built and when it was sold.

```python
## list of temporal variables
year_feature=[feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
## lets look the contxt of these year variables
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

### Analyse the temporal variables 

```python
## lets find out the relation(if any) between house sold in 
## particular year and the sales price
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

from the above observation we find that with the increase in year the sales price is decreasing

```python
year_feature
```

```python
## here we will find the difference between allyear feature with respect to sale price
for feature in year_feature:
    if feature != 'YrSold':
        data=dataset.copy()
        # here we will calculate the d/f b/w all year feature with Year Sold
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

```python
## Numerical variables are of 2 types
## 1. Continous
## 2. Discrete(fixed set of integer values)

```

##### Discrete Variables

```python
discret_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('Discrete Variables {}',format(len(discret_feature)))
```

```python
discret_feature
```

```python
dataset[discret_feature].head()
```

```python
## find therelationship b/w discrete features and Sals Price

for feature in discret_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.title(feature)
    plt.show()
```

##### Continous Variables

```python
continous_feature=[feature for feature in numerical_features if feature not in discret_feature + year_feature+['Id']]
print('Continous feature count is', format(len(continous_feature)))
```

```python
## now analyse the continous feature using histogram to find the distribution
for feature in continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
## we will use lograthmic transformation

for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
## lets find outthe outliers
for feature in continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
